Since all the columns with missing values are categorical, mode will be used to impute the missing values


a) Imputing Race column with the value that has the highest count, which is the mode, and confirming that it was successful

```python


raceValue=pharma_df['Race'].value_counts().index[0]
#pharma_df.loc[(pharma_df['col1']==val1) & (df['col2'].isnull()), 'col2'] = replacement_value
pharma_df['Race'].fillna(value=raceValue, inplace=True)


```

```python
pharma_df.tail(50)
```

b) Imputing Ethnicity column with value that has the highest count, which is the mode

```python

ethValue=pharma_df['Ethnicity'].value_counts().index[0]
#pharma_df.loc[(pharma_df['col1']==val1) & (df['col2'].isnull()), 'col2'] = replacement_value
pharma_df['Ethnicity'].fillna(value=ethValue, inplace=True)
```

c) Imputing 'Region' column with value that has the highest count, which is the mode and checking whether it is successful

```python

regValue=pharma_df['Region'].value_counts().index[0]
#pharma_df.loc[(pharma_df['col1']==val1) & (df['col2'].isnull()), 'col2'] = replacement_value
pharma_df['Region'].fillna(value=regValue, inplace=True)
```

d) Imputing 'Ntm_speciality' column with value that has the highest count, which is the mode

```python

specValue=pharma_df['Ntm_Speciality'].value_counts().index[0]
#pharma_df.loc[(pharma_df['col1']==val1) & (df['col2'].isnull()), 'col2'] = replacement_value
pharma_df['Ntm_Speciality'].fillna(value=specValue, inplace=True)
```

Filling in the missing values of "Risk_Segment_During_Rx", "Tscore_Bucket_During_Rx","Change_T_Score","Change_Risk_Segment" columns .This can be done jointly as below

```python
cols = ["Risk_Segment_During_Rx", "Tscore_Bucket_During_Rx","Change_T_Score","Change_Risk_Segment"]
pharma_df[cols]=pharma_df[cols].fillna(pharma_df.mode().iloc[0])
```
